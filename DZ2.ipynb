Напишите функцию, которая принимает в качестве аргумента номер студента и новый номер группы и позволяет изменить номер группы этого студента на новый внутри созданного словаря.
​
Напишите функцию, которая принимает номер группы и выводит список (фамилии, имена, отчества) всех студентов этой группы. Необходимо обращаться к словарю.

```python
def get_all_stud_dictionary(group_name):
    global all_stud_dictionary
    for _, fitch in all_stud_dictionary.items():
        if (fitch[-1] == group_name):
            for elem in fitch[0:3]:
                print(elem, end=' ')
            print()


def all_stud_dictionary_change_group(student_id, new_group_name):
    global all_stud_dictionary
    if student_id in all_stud_dictionary:
        all_stud_dictionary[student_id][-1] = new_group_name


students = [
    ['0001', 'Антонов', 'Антон',
     'Игоревич', '20.08.2009', 'БСТ161'],
    ['1102', 'Богов', 'Артем',
     'Игоревич', '25.01.2010', 'БСТ162'],
    ['0333', 'Глаголева', 'Анастасия',
     'Николаевна', '11.07.2009', 'БСТ163'],
    ['4004', 'Степанова', 'Наталья',
     'Александровна', '13.02.2008', 'БСТ161'],
    ['0045', 'Боков', 'Игорь',
     'Харитонович', '02.06.2009', 'БСТ161'],
    ['0096', 'Васильков', 'Валентин',
     'Сергеевич', '20.03.2009', 'БСТ164'],
    ['0607', 'Сиропова', 'Виолетта',
     'Эдуардовна', '28.05.2010', 'БСТ162'],
]

all_stud_dictionary = {}
for stud in students:
    all_stud_dictionary[stud[0]] = stud[1:]
print('словарь из списка студентов:\n')
print(all_stud_dictionary)
```
